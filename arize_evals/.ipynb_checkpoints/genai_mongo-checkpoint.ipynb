```python
mpl.rcParams['font.size'] = 10.0

def plot_pie_chart(df, column, fsize=(5,5)):
    t = df.groupby([column]).size().sort_values(ascending=False)
    ax = t.plot(kind='pie', title=column, figsize=fsize, autopct='%1.1f%%')
    plt.show()
    print(t.to_string())
    
def groupby(df, columns):
    return df.groupby(columns).size().reset_index(name='counts').sort_values('counts', ascending=False).reset_index(drop=True)

def show_accumulated_count(df, groupByCols, timeCol, title):
    t = df.groupby(groupByCols).agg({timeCol: ['min']}).reset_index()
    t.columns = groupByCols + ['time']
    t = t.sort_values(by=['time'], ascending=[True]).reset_index(drop=True)
    mask = t.time.isnull()
    t = pd.concat([t[mask], t[~mask]]).reset_index(drop=True)
    t['count'] = t.index + 1
    t.set_index('time')['count'].plot(title=title)
```

```python
def plot_pie_chart2(df, columns, fsize=(5,5)):
    t = df.groupby(columns).size().sort_values(ascending=False)
    ax = t.plot(kind='pie', title=' + '.join(columns), figsize=fsize, autopct='%1.1f%%')
    plt.show()
    print(t.to_string())
```

```python
def percentage(n, total):
    if (n == 0):
        return 0.0
    else:
        return 100.0 * n / total
        
def summary(messages):
    total = len(messages['session_id'].unique()) 
    summary = len(messages[~messages['parsed.results.summary'].isna()]['session_id'].unique())
    perc = (100.0 * summary/total)
    print('   With session count = %s, number of sessions with summary = %s -> perc = %s' % (total, summary, perc))
    ready = len(messages[~messages['parsed.results.summary'].isna() & messages['parsed.results.ready_for_therapy']])
    not_ready = len(messages[~messages['parsed.results.summary'].isna() & (messages['parsed.results.ready_for_therapy'] == False)])
    print('   Number of sessions with summary & ready for therapy =', ready, '-> perc =', percentage(ready, summary))
    print('   Number of sessions with summary & NOT ready for therapy =', not_ready, 'perc % =', percentage(not_ready,summary))
```
